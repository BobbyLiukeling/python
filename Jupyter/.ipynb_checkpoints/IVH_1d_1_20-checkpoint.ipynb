```python
import stlearn as st
import scanpy
import tables
st.settings.set_figure_params(dpi=120)
# Reading data
string = "C:\\Users\\Jaydar1219\\anaconda3\\envs\\stlearn\\stlearn_demo\\"
paths = ["1-20","1-50"]
for path in paths:
    path = string+path
    pass
data = st.Read10X(path="C:\\Users\\Jaydar1219\\anaconda3\\envs\\stlearn\\stlearn_demo\\IVH_1d_1_20")
# Save raw_count
data.layers["raw_count"] = data.X
# Preprocessing
st.pp.filter_genes(data,min_cells=3)
st.pp.normalize_total(data)
st.pp.log1p(data)
# Keep raw data
data.raw = data
st.pp.scale(data)
```

```python
# Run PCA
st.em.run_pca(data,n_comps=50,random_state=0)
# Tiling image
st.pp.tiling(data,out_path="tiling",crop_size = 40)
# Using Deep Learning to extract feature
st.pp.extract_feature(data)
# Apply stSME spatial-PCA option
st.spatial.morphology.adjust(data,use_data="X_pca",radius=50,method="mean")
st.pp.neighbors(data,n_neighbors=25,use_rep='X_pca_morphology',random_state=0)
st.tl.clustering.louvain(data,random_state=0)
```

```python
# annote them with names
data.obs['louvain_anno'] = data.obs['louvain']
data.obs['louvain_anno'].cat.categories = ['0/cortex layer V,Ⅵa,Ⅵb', '1/Piriform area', '2/Midbrain', '3/Thalamus', '4/Hypothalamic lateral zone', '5/Field CA1 of the hippocampus', '6/Hypothalamic nucleus', '7/Fornix system', '8/cortex layer Ⅱ,Ⅲ,Ⅳ', '9/Field CA3 of the hippocampus', '10/Cortical amygdalar area', '11/Retrosplenial area', '12/Lateral ventricle',
       '13/Hippocampal formation', '14/cerebal peduncle', '15/Cortex layer I', '16/Dentate gyrus']
```

```python
st.pl.cluster_plot(data,use_label="louvain_anno",tissue_alpha=1,spot_size=7,show_legend=True)
```
